```python
#Compress the dataframe

def compress(x, thresh=1):
    n = len(x)
    i, j = 0, 1
    idxs = [i]

    while j < n:
        if abs(x[i] - x[j]) < thresh:
            j += 1
        else:
            idxs.append(j)
            i = j
            j += 1

    return idxs
```
